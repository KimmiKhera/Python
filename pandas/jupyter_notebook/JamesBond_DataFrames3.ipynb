## A Review of the .apply() Method on Single Columns:

```python
bond = pd.read_csv("C:/Users/kkher/Desktop/Pandas/jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
## apply method applies some kind of operation to every value within the series. it is similar to a broadcasting operation such as multiplying every value or dividing every value.
```

```python
def convert_to_string_and_add_millions(number):
    return str(number) + " MILLIONS!"
```

```python
columns = ["Box Office", "Budget", "Bond Actor Salary"]
for col in columns:
    bond[col] = bond[col].apply(convert_to_string_and_add_millions)
```

```python
bond.head()
```

## The .apply() Method with Row Values:

```python
bond = pd.read_csv("C:/Users/kkher/Desktop/Pandas/jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
def good_movie(row):
    
    actor = row[1]
    budget = row[4]
    
    if actor == "Pierce Brosnan":
        return "The best"
    elif actor == "Roger Moore" and budget > 40:
        return "Enjoyable"
    else:
        return "I have no clue"
bond.apply(good_movie, axis ="columns")
```
